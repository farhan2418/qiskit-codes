```python
import matplotlib.pyplot as plt
import numpy as np
data = np.array([3, 6, 9, 12, 15])
values = np.array([58, 110, 170, 230, 280])
slope, intercept = np.polyfit(data, values, 1)
line = int(slope) * data + intercept
print(slope)
# plt.plot(data, slope * data + intercept, color='red', label=f'Fit line (slope={slope:.2f})')
plt.scatter(data, values, color='green')
plt.plot(data, line, color='red')
plt.grid(True)
plt.xticks(np.arange(0, 16, 3))
plt.yticks(np.arange(0, 300, 60))

plt.xlabel('fringes counted')
plt.ylabel('change in pressure')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

# Fit a line (1st degree polynomial)
slope, intercept = np.polyfit(x, y, 1)

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Generate a line based on the fit
line = slope * x + intercept

# Plot the original data and the fit line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, line, color='red', label='Best Fit Line')
plt.title("Line Fit")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Set major ticks
plt.xticks(np.arange(0, 11, 1))  # Major ticks on x-axis every 1 unit
plt.yticks(np.arange(-1, 1.5, 0.5))  # Major ticks on y-axis every 0.5 units

# Optionally set minor ticks
from matplotlib.ticker import AutoMinorLocator
ax = plt.gca()
ax.xaxis.set_minor_locator(AutoMinorLocator(2))  # 2 minor ticks between each major tick
ax.yaxis.set_minor_locator(AutoMinorLocator(2))

plt.grid(which='both')  # Enable grid for both major and minor ticks

plt.show()
```
