```python
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Statevector, Operator, partial_trace
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram, array_to_latex
from qiskit.quantum_info import Statevector, DensityMatrix
```

```python
qc = QuantumCircuit(4)
for i in range(4):
    qc.h(i)
qc.cz(0,1)
qc.cz(1,2)
qc.cz(1,3)
qc.draw("mpl")
```

```python
psi = Statevector(qc)
psi.draw("latex")
```

```python
rho = DensityMatrix(psi)
rho.draw("latex")
```

```python
unitary = Operator(qc)
unitary.draw("latex")
```

```python
import numpy as np
qc1 = QuantumCircuit(2)
qc1.x(0)
qc1.ry(np.pi/3,0)
qc1.cx(0,1)
qc1.draw("mpl")
```

```python
psi = Statevector(qc1)
psi.draw("bloch")
```

```python
rho2 = DensityMatrix(qc1)
rhoa = partial_trace(rho2,[0])
rhob = partial_trace(rho2,[1])
rhoa.draw("latex")
```
